````python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator

def fit_LSTM(
    scaled_train_data,
    time_steps,
    n_features,
    batch_size=6,
    activation='relu',
    epochs=10,
    verbose=0
):
    generator = TimeseriesGenerator(
        data=scaled_train_data,
        targets=scaled_train_data,
        length=time_steps,
        batch_size=batch_size
    )

    model = Sequential()
    model.add(LSTM(200, activation=activation,
                   input_shape=(time_steps, n_features)))
    model.add(Dropout(0.25))
    model.add(Dense(1))

    model.compile(optimizer='adam', loss='mse')
    model.fit(generator, epochs=epochs, verbose=verbose)

    return model
````

````python
def predict_LSTM(
    scaled_train_data,
    scaler,
    train_data,
    test_data,
    time_steps,
    n_features,
    lstm_model
):
    preds = []

    batch = scaled_train_data[-time_steps:].reshape(
        (1, time_steps, n_features)
    )

    for i in range(len(test_data)):
        pred = lstm_model.predict(batch, verbose=0)[0]
        preds.append(pred)
        batch = np.append(batch[:, 1:, :], [[pred]], axis=1)

    preds = scaler.inverse_transform(preds)

    df_pred = pd.DataFrame(
        preds,
        index=test_data.index,
        columns=['LSTM_Pred']
    )

    # Plot (để so sánh SARIMA)
    plt.figure(figsize=(12,5))
    plt.plot(train_data, label='Train')
    plt.plot(test_data, label='Test')
    plt.plot(df_pred, label='LSTM Forecast')
    plt.legend()
    plt.title('LSTM Forecast – PM10 (Weekly)')
    plt.show()

    return df_pred
````

````python
import torch
import torch.nn as nn
import torch.nn.functional as F

# Define the Model class for iTransformer based on common implementations
class Model(nn.Module):
    def __init__(self, configs):
        super(Model, self).__init__()
        self.seq_len = configs.seq_len
        self.pred_len = configs.pred_len
        self.model_dim = configs.d_model
        self.num_heads = configs.n_heads
        self.num_encoder_layers = configs.e_layers
        self.dim_feedforward = configs.d_ff
        self.dropout = configs.dropout
        self.activation = F.gelu

        # Inverted Transformer: encode features, not time steps
        self.encoder_layer = nn.TransformerEncoderLayer(
            d_model=self.model_dim,
            nhead=self.num_heads,
            dim_feedforward=self.dim_feedforward,
            dropout=self.dropout,
            activation=self.activation,
            batch_first=True  # Input and output tensors are (batch, sequence, feature)
        )
        self.transformer_encoder = nn.TransformerEncoder(
            self.encoder_layer,
            num_layers=self.num_encoder_layers
        )

        # Linear projection to output dimension (pred_len)
        self.projection = nn.Linear(self.model_dim, self.pred_len)

    def forward(self, x_enc):
        # x_enc: [Batch, Input_sequence_length, N_Features]
        # For iTransformer, we transpose: [Batch, N_Features, Input_sequence_length]
        # And then feed N_Features as the sequence length to the Transformer Encoder.
        # The 'features' in a typical transformer become 'time steps' for iTransformer.
        # The 'time steps' in a typical transformer become 'features' for iTransformer.

        # x_enc shape is [Batch, Sequence_Length, Num_Features]
        # iTransformer transposes this to [Batch, Num_Features, Sequence_Length]
        # where Num_Features acts as the 'sequence' for the Transformer encoder
        # and Sequence_Length acts as the 'feature dimension' for each 'token'

        # Original paper's logic often involves a projection first or handling feature dimension
        # Here, I will adapt the input to fit a standard TransformerEncoder as closely as possible
        # by treating each feature as a token in a sequence.

        # However, the most direct interpretation of iTransformer is to invert the dimensions,
        # where features become the 'sequence' for the transformer, and time steps become the 'feature dimension' of each 'token'.
        # This means the input to the transformer encoder should be [Batch, N_Features, Seq_Len]

        # Let's adjust for the standard TransformerEncoder which expects [Batch, Sequence_Length, Embedding_Dimension]
        # If we treat each feature as a 'token' in a sequence, then N_Features is the Sequence_Length
        # and the original Sequence_Length of the time series becomes the 'Embedding_Dimension' for these 'tokens'.

        # Initial projection to d_model for the time dimension
        x_enc_proj = x_enc.permute(0, 2, 1) # [Batch, N_Features, Seq_Len]
        # Now, each feature (N_Features) is a sequence item, and its value across time (Seq_Len) is its embedding.
        # We need to project Seq_Len to d_model if Seq_Len != d_model
        if x_enc_proj.shape[2] != self.model_dim:
            self.feature_embedding = nn.Linear(self.seq_len, self.model_dim).to(x_enc.device)
            x_enc_proj = self.feature_embedding(x_enc_proj) # [Batch, N_Features, d_model]

        # Transformer Encoder expects [Batch, Sequence_Length, d_model]
        # Here, Sequence_Length is N_Features, and d_model is our target embedding dimension
        enc_output = self.transformer_encoder(x_enc_proj) # [Batch, N_Features, d_model]

        # After encoding, we want to predict future time steps for each feature.
        # We need to project d_model back to pred_len for each feature
        # Then transpose back to original series shape [Batch, Pred_Len, N_Features]
        predictions = self.projection(enc_output) # [Batch, N_Features, Pred_Len]

        # Final transpose to get [Batch, Pred_Len, N_Features] (standard output format for multivariate TS forecasting)
        predictions = predictions.permute(0, 2, 1) # [Batch, Pred_Len, N_Features]

        # If it's a univariate series, squeeze the last dimension back to [Batch, Pred_Len]
        if predictions.shape[2] == 1:
            predictions = predictions.squeeze(-1)

        return predictions

print("iTransformer Model class re-defined successfully.")
````

````python
import torch
import torch.nn as nn

class MovingAverageBlock(nn.Module):
    def __init__(self, kernel_size):
        super(MovingAverageBlock, self).__init__()
        self.kernel_size = kernel_size
        self.avg_pool = nn.AvgPool1d(kernel_size=kernel_size, stride=1, padding=kernel_size // 2, count_include_pad=False)

    def forward(self, x):
        # x: [Batch, Input_sequence_length, N_Features]
        # AvgPool1d expects [Batch, N_Features, Input_sequence_length]
        x_permuted = x.permute(0, 2, 1)
        seasonal = x_permuted - self.avg_pool(x_permuted)
        return seasonal.permute(0, 2, 1)

class DLinear(nn.Module):
    def __init__(self, seq_len, pred_len, n_features, kernel_size=25):
        super(DLinear, self).__init__()
        self.seq_len = seq_len
        self.pred_len = pred_len
        self.n_features = n_features

        self.moving_average = MovingAverageBlock(kernel_size)

        # Linear layer for the seasonal component
        self.seasonal_linear = nn.Linear(seq_len, pred_len)

        # Linear layer for the trend component
        self.trend_linear = nn.Linear(seq_len, pred_len)

    def forward(self, x):
        # x: [Batch, Input_sequence_length, N_Features]

        # Decomposition: x is input for prediction, x_series is target (not used for seasonal/trend calc)
        # For decomposition, we need to decompose x. The output of moving_average will be seasonal part.
        # The trend part is the moving average itself.
        # Note: The original DLinear uses a simple moving average for decomposition, which is different from a seasonal block.
        # This implementation aligns with the paper's concept of decomposition into seasonal and trend.

        x_permuted = x.permute(0, 2, 1) # [Batch, N_Features, Input_sequence_length]
        seasonal_part_input = x_permuted - self.moving_average.avg_pool(x_permuted)
        trend_part_input = self.moving_average.avg_pool(x_permuted)

        # Seasonal prediction
        seasonal_output = self.seasonal_linear(seasonal_part_input)

        # Trend prediction
        trend_output = self.trend_linear(trend_part_input)

        # Combine seasonal and trend predictions
        output = (seasonal_output + trend_output).permute(0, 2, 1) # [Batch, Pred_len, N_Features]

        return output

print("DLinear model and MovingAverageBlock classes defined successfully.")
````

````python
from sklearn.preprocessing import StandardScaler

# 1. Extract the 'PM10' column from train_wk and test_wk DataFrames as NumPy arrays and reshape.
train_wk_pm10 = train_wk.PM10.values.reshape(-1, 1)
test_wk_pm10 = test_wk.PM10.values.reshape(-1, 1)

# 2. Initialize a StandardScaler instance.
scaler_patchtst = StandardScaler()

# 3. Fit the StandardScaler on the reshaped training data and then transform both train and test data.
scaled_train_patchtst = scaler_patchtst.fit_transform(train_wk_pm10)
scaled_test_patchtst = scaler_patchtst.transform(test_wk_pm10)

# 4. Use the create_dataset function to generate X_train_patchtst and y_train_patchtst.
X_train_patchtst, y_train_patchtst = create_dataset(scaled_train_patchtst, seq_len, pred_len)

# 5. Create X_predict_patchtst for forecasting the test_wk period.
X_predict_patchtst = scaled_train_patchtst[-seq_len:].reshape((1, seq_len, n_features))

# 6. Print the shapes of the created datasets to verify their structure.
print(f"Shape of train_wk_pm10: {train_wk_pm10.shape}")
print(f"Shape of test_wk_pm10: {test_wk_pm10.shape}")
print(f"Shape of scaled_train_patchtst: {scaled_train_patchtst.shape}")
print(f"Shape of scaled_test_patchtst: {scaled_test_patchtst.shape}")
print(f"Shape of X_train_patchtst: {X_train_patchtst.shape}")
print(f"Shape of y_train_patchtst: {y_train_patchtst.shape}")
print(f"Shape of X_predict_patchtst: {X_predict_patchtst.shape}")
````

````python
import torch
import torch.nn as nn
import torch.nn.functional as F

# 2. Update the Configs class
class Configs:
    seq_len = seq_len
    pred_len = pred_len
    d_model = 64
    n_heads = 4
    e_layers = 2
    d_ff = 128
    factor = 5
    dropout = 0.1
    activation = 'gelu'
    embed = 'fixed'
    freq = 'w'
    output_attention = False
    use_norm = True
    class_strategy = None

    # New parameters for PatchTST
    patch_len = 4
    stride = 4

# 3. Define a Patching class
class Patching(nn.Module):
    def __init__(self, seq_len, patch_len, stride):
        super(Patching, self).__init__()
        self.seq_len = seq_len
        self.patch_len = patch_len
        self.stride = stride

    def forward(self, x): # x: [Batch, Sequence_Length, N_Features]
        n_features = x.shape[-1]

        # Permute to [Batch, N_Features, Sequence_Length] for padding
        x_permuted = x.permute(0, 2, 1)

        # Determine padding amount
        L = self.seq_len
        P = self.patch_len
        S = self.stride

        remainder = (L - P) % S
        if remainder != 0:
            padding_needed = S - remainder
        else:
            padding_needed = 0

        # Pad the sequence dimension
        if padding_needed > 0:
            x_padded = F.pad(x_permuted, (0, padding_needed), 'replicate')
            padded_seq_len = x_padded.shape[-1]
        else:
            x_padded = x_permuted
            padded_seq_len = L

        # Unfold to create patches
        # unfold operates on the last dimension (sequence length)
        patches = x_padded.unfold(dimension=-1, size=self.patch_len, step=self.stride)
        # patches shape: [Batch, N_Features, N_patches, patch_len]

        # Reshape to [Batch, N_patches, patch_len * N_Features]
        patches = patches.permute(0, 2, 1, 3).reshape(
            x.shape[0], patches.shape[-2], self.patch_len * n_features
        )

        return patches, padded_seq_len

# 4. Define a PatchEmbedding class
class PatchEmbedding(nn.Module):
    def __init__(self, d_model, patch_len, n_features, dropout):
        super(PatchEmbedding, self).__init__()
        self.patch_lin = nn.Linear(patch_len * n_features, d_model)
        self.dropout = nn.Dropout(dropout)

    def forward(self, patches): # patches: [Batch, N_patches, patch_len * N_Features]
        embedded_patches = self.patch_lin(patches)
        embedded_patches = self.dropout(embedded_patches)
        return embedded_patches

# 5. Define the main PatchTST model class
class Model(nn.Module):
    def __init__(self, configs):
        super(Model, self).__init__()
        self.seq_len = configs.seq_len
        self.pred_len = configs.pred_len
        self.n_features = 1 # Assuming univariate for now, will generalize if needed
        self.d_model = configs.d_model
        self.patch_len = configs.patch_len
        self.stride = configs.stride

        self.patching_layer = Patching(self.seq_len, self.patch_len, self.stride)

        # Calculate num_patches dynamically (pass dummy tensor for shape inference)
        dummy_x = torch.zeros(1, self.seq_len, self.n_features)
        dummy_patches, _ = self.patching_layer(dummy_x)
        self.num_patches = dummy_patches.shape[1]

        self.patch_embedding = PatchEmbedding(
            self.d_model, self.patch_len, self.n_features, configs.dropout
        )

        encoder_layer = nn.TransformerEncoderLayer(
            d_model=self.d_model,
            nhead=configs.n_heads,
            dim_feedforward=configs.d_ff,
            dropout=configs.dropout,
            activation=configs.activation,
            batch_first=True
        )
        self.transformer_encoder = nn.TransformerEncoder(
            encoder_layer,
            num_layers=configs.e_layers
        )

        self.head = nn.Linear(self.num_patches * self.d_model, self.pred_len * self.n_features)

    def forward(self, x_enc): # x_enc: [Batch, Sequence_Length, N_Features]
        # 1. Patching
        patches, _ = self.patching_layer(x_enc)
        # patches: [Batch, N_patches, patch_len * N_Features]

        # 2. Patch Embedding
        embedded_patches = self.patch_embedding(patches)
        # embedded_patches: [Batch, N_patches, d_model]

        # 3. Transformer Encoder
        transformer_output = self.transformer_encoder(embedded_patches)
        # transformer_output: [Batch, N_patches, d_model]

        # 4. Flatten and Head Layer
        flattened_output = transformer_output.view(
            transformer_output.size(0), -1
        ) # [Batch, N_patches * d_model]

        predictions = self.head(flattened_output)
        # predictions: [Batch, pred_len * n_features]

        # 5. Reshape output
        predictions = predictions.view(
            x_enc.size(0), self.pred_len, self.n_features
        )

        # If univariate, squeeze the last dimension
        if self.n_features == 1:
            predictions = predictions.squeeze(-1)

        return predictions

print("PatchTST Model architecture (Patching, PatchEmbedding, and Model classes) defined successfully.")
````
